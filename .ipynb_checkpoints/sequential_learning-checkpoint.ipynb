```python
import warnings
warnings.simplefilter('ignore')

# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame

exoplanets = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
exoplanets = exoplanets.dropna(axis='columns', how='all')
# Drop the null rows
exoplanets = exoplanets.dropna()
exoplanets.head(10)
```

```python
# Set features. This will also be used as your x values.
selected_features = exoplanets[['koi_disposition', 'koi_impact', 'koi_duration', 'koi_depth', 'koi_prad','koi_teq','koi_insol', 'koi_model_snr','koi_slogg']]
selected_features['koi_disposition']=selected_features['koi_disposition'].map({'CONFIRMED': 1, 'CANDIDATE': 2, 'FALSE POSITIVE':3})
```

```python
# Create a Train Test Split

# Use `koi_disposition` for the y values
```

```python
X = selected_features.drop("koi_disposition", axis=1)
# y = selected_features["koi_disposition"]
y = selected_features['koi_disposition']
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
print(X_train.shape,y_train.shape)
```

# Pre-processing
#Scale the data using the MinMaxScaler and perform some feature selection

```python
# Import dependencies
from sklearn.preprocessing import LabelEncoder, MinMaxScaler,StandardScaler
from tensorflow.keras.utils import to_categorical

# scale the data
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python

# Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)

print(y_train_categorical)
```

# Train the Model

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models, utils

# Create a Keras model that's compatible with scikit-learn
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

import numpy as np
import sklearn
import sklearn.datasets

import matplotlib.pyplot as plt

from numpy.random import seed
seed(1)
```

```python
classifier = models.Sequential()
classifier.add(layers.Dense(units=100, activation='relu', input_dim=8))
classifier.add(layers.Dense(units=100, activation='relu'))
classifier.add(layers.Dense(units=100, activation='relu'))
classifier.add(layers.Dense(units=3, activation='softmax'))
# classifier.summary()
classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
classifier.summary()
```

```python
# callbacks = [EarlyStopping(monitor='loss', patience=2)]
classifier.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = classifier.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = classifier.predict_classes(X_test_scaled[:10])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:10])}")
```
